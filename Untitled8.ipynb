```python
import pandas as pd
data={
    "Posion":["HR","IT","HR","IT","MARKAI"],
    "Age":[20,21,18,21,22],
    "Sal":[1000000,1000000,130000,90000,1200000]
}
df=pd.DataFrame(data)
print(df)
```

```python
Same_Pos=df[df["Posion"]=="HR"]
print(Same_Pos)
```

```python
high_sal=df[df["Sal"]>100000]
print(high_sal)
```

```python
sorted_df=df.sort_values("Sal")
print(sorted_df)
```

```python
import pandas as pd

data = {
    "Region": ["South","North","East","West","South","North"],
    "Product":
    ["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales": [50000,30000,45000,25000,35000,40000]

}
df=pd.DataFrame(data)
print(df)

```

```python
south_sales = df[df["Region"] == "South"]

print(south_sales)
```

```python
high_sales = df[df["Sales"] > 30000]

print(high_sales)
```

```python
sorted_df = df.sort_values("Sales")

print(sorted_df)
```
