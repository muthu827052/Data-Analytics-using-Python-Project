```python
import pandas as pd
data = {
    "Region":["South","North","East","West","South","North"],
    "Product":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales":[50000,30000,45000,20000,35000,40000]
}
df = pd.DataFrame(data)
print(df)
```

DAY 7

```python
Region_sales = df.groupby("Region")["Sales"].sum()

print(Region_sales)
```

```python
print(df["Sales"].sum())
```

```python
product_sales = df.groupby("Product")["Sales"].sum()

print(product_sales)
```

```python
result = df.groupby("Region")["Sales"].agg(
    ["sum", "mean", "max", "min"]
)
print(result)
```

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]

sales = [10000,12000,15000,13000,18000]

plt.plot(months,sales,
         color="green",
         marker="o",
         linestyle="--")

plt.title("Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.show()
```

CLEAN A MESSY SALES DATASET

```python
import pandas as pd
import numpy as np

data = {
    "Product": ["Laptop", "Mobile", "Tablet", None, "Laptop"],
    "Price": [50000, 20000, None, 30000, 50000],
    "Quantity": [2, 3, 1, None, 2]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

df = df.drop_duplicates()

df["Price"] = df["Price"].fillna(df["Price"].mean())

df["Quantity"] = df["Quantity"].fillna(0)

print("\nCleaned Dataset:")
print(df)
```
