PIE CHART

```python
import matplotlib.pyplot as plt

expenses = [5000,3000,2000]

categories = ["Food","Travel","Shoping"]

plt.pie(expenses,
        labels=categories,autopct="%1.1f%%")
plt.show()
```

Monthly Sales Dashboard

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun"]

sales = [10000,12000,15000,13000,18000,20000]

plt.figure(figsize=(18,5)) # Increased figsize to accommodate 3 plots horizontally

#Line Chart
plt.subplot(1,3,1)

plt.plot(months,
       sales,
         marker="o")

plt.title("Sales Trend") # Added title to line chart

plt.xlabel("Months")

plt.ylabel("Sales")

#Bar Chart
plt.subplot(1,3,2)

plt.bar(months,
        sales,
        color="orange")

plt.title("Sales Comparison")

plt.xlabel("Months")

plt.ylabel("Sales")

#Pie Chart
plt.subplot(1,3,3)

expenses = [5000,3000,2000]

categories = ["Food","Travel","Shoping"]

plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%",colors = ["pink","yellow","green"])

plt.title("Expense Distribution")

plt.tight_layout()

plt.show()
```

Distrogram

```python
import matplotlib.pyplot as plt

marks = [45,50,55,60,65,70,72,75,80,85,90,95]

plt.hist(marks, bins=5)

plt.title("Students Marks Distribution")

plt.xlabel("Marks")

plt.ylabel("Number of Students")

plt.show()
```

Hours vs Marks

```python
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7,8]

marks = [45,50,55,60,65,70,72,75]

plt.scatter(hours,marks)

plt.title("Hours vs Marks")

plt.xlabel("Hours")

plt.ylabel("Marks")
plt.show()
```

```python
import pandas as pd

# Example DataFrame definition (you would replace this with your actual data)
data = {
    'Region': ['East', 'West', 'East', 'West', 'North', 'South', 'North'],
    'Sales': [100, 150, 0, 200, 50, 120, 0]
}
df = pd.DataFrame(data)

df_active = df[df['Sales'] > 0]

df_sorted = df_active.sort_values(
    by=['Region','Sales'],
    ascending=[True,False]
)

report = df_sorted.groupby('Region')['Sales'].agg(
    ['sum','mean','count']
)

# Rename Columns
report.columns = ['Total Sale','Avg Sale','Order Count']

print(report)
```
