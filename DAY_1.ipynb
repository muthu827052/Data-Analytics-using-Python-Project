```python
marks =[85,92,78,90,65]

total_marks =sum(marks)
average = total_marks/len(marks)

print("Total marks:",total_marks)
print("Average marks:",average)

if average >= 90:
    print("Grade: A")
elif average >= 80:
    print("Grade: B")
elif average >= 70:
    print("Grade: C")
elif average >= 60:
    print("Grade: D")
else:
    print("Grade: Fail")
```
